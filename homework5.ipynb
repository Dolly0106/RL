```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

# 創建一個Service對象並指定chromedriver的路徑
s=Service('D:\\chromedriver\\chromedriver.exe')

# 將Service對象傳遞給webdriver.Chrome()
browser = webdriver.Chrome(service=s)
```

```python
#網頁
url = 'https://www.technice.com.tw/techmanage/85107/'
browser.get(url)
print(browser.page_source)
```

```python
from selenium.webdriver.common.by import By
elements = browser.find_elements(By.CLASS_NAME, "post-title.entry-title")# 抓取所有
for element in elements:
    print(element.text)
```

```python
#報導內容
element_xpath=browser.find_element(By.XPATH,"html/body/div[1]/div/div/div/div[1]/div/article/div[1]")
print(element_xpath.text)
```

```python
import csv
# 建立一個空的列表來儲存資料
data = []
text = element_xpath.text

  # 將文字屬性添加到資料列表中
data.append({'text': text})
```

```python
data
```

```python
import jieba
import collections
import matplotlib.pyplot as plt

# 文字分割
text = data[0]['text']
seg_list = jieba.cut(text, cut_all=False)

# 計算詞頻
word_counts = collections.Counter(seg_list)
word_counts_top10 = word_counts.most_common(10)

# 繪製長條圖
words, counts = zip(*word_counts_top10)
plt.bar(words, counts)
plt.show()
```

```python
import jieba
import collections
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 指定字體路徑
myfont = FontProperties(fname='C:\Windows\Fonts\kaiu.ttf')

# 文字分割
text = data[0]['text']
seg_list = jieba.cut(text, cut_all=False)

# 計算詞頻
word_counts = collections.Counter(seg_list)

word_counts_top10 = word_counts.most_common(10)

# 繪製長條圖
words, counts = zip(*word_counts_top10)
plt.bar(range(len(words)), counts)

# 設定標籤，並指定字體
plt.xticks(range(len(words)), words, fontproperties=myfont)

# 顯示圖表
plt.show()
```
